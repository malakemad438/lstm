```python
import math

# activation functions
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

# weights
Wf, Whf = 0.5, 0.1
Wi, Whi = 0.6, 0.2
Wc, Whc = 0.7, 0.3
Wo, Who = 0.8, 0.4

bf = bi = bc = bo = 0

# prediction layer
Wy = 4
by = 0

# initial states
h = 0
c = 0

# input sequence
inputs = [1,2,3,4]

print("LSTM Example\n")

for t,x in enumerate(inputs,1):

    print("t =",t,"  x =",x)

    # forget gate
    ft = sigmoid(Wf*x + Whf*h + bf)
    print("f_t =",round(ft,3))

    # input gate
    it = sigmoid(Wi*x + Whi*h + bi)
    print("i_t =",round(it,3))

    # candidate
    c_tilde = tanh(Wc*x + Whc*h + bc)
    print("c_tilde =",round(c_tilde,3))

    # update cell state
    c = ft*c + it*c_tilde
    print("c_t =",round(c,3))

    # output gate
    ot = sigmoid(Wo*x + Who*h + bo)
    print("o_t =",round(ot,3))

    # hidden state
    h = ot * tanh(c)
    print("h_t =",round(h,3))

    print("------------------")

# prediction
y_hat = Wy*h + by
print("prediction =",round(y_hat,2))
```
